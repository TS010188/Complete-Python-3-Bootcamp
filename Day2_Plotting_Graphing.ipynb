```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Pandas, Seaborn and Plotly
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(2)
```

```python
titanic.info()
```

```python
# Bar Plot using mathlotlib
plt.bar(titanic['class'].unique(),titanic['class'].value_counts() )
plt.xlabel("Class of Tickets on Titanic")
plt.ylabel("Count")
plt.title("Count by Class on the Titanic")
```

```python
# Count plot with seaborn
fig = sns.countplot(x="class", data=titanic)
plt.xlabel("Class of Tickets on Titanic")
plt.ylabel("Count")
plt.title("Count by Class on the Titanic")
plt.show(fig)
```

```python
# Bar Plot plotly
fig3 = px.bar(titanic,x="class",title="Count by Class on the Titanic",color ="class", labels={"class": "Class of Tickets on Titanic"} )
fig3.show()
```

```python
# Histogram with Plotly
fig4 = px.histogram(titanic,x="fare",title="Histogram of Fares on the Titanic",labels={"fare": "Fare"},color_discrete_sequence = px.colors.qualitative.Set1 )
fig4.show()
```

```python
# Histogram by survived with plotly
fig5 = px.histogram(titanic,x="age",title="Histogram of Ages by Survival",labels={"age": "Age"},color_discrete_sequence = px.colors.qualitative.Dark24, color = "survived" )
fig5.show()
```

```python
# Histogram with marplotlib
plt.hist(titanic['fare'],color="lightblue",alpha=0.6,edgecolor = "black",bins=30 )
plt.xlabel("Fare")
plt.ylabel("Count")
plt.figure(figsize=(12,5))
```

```python
gapminder = px.data.gapminder()
```

```python
gapminder.head(5)
```

```python
gapminder.info()
```

```python
gapminder.continent.unique()
```

```python
plt.hist(gapminder['lifeExp'], edgecolor = 'black', bins = 15)
plt.title('Histogram of Life Expectancy')
```

```python
# Density plot
# A Kernel Density Estimate Plot
sns.displot(data = gapminder, x = 'lifeExp', kind = 'kde', hue = 'continent',
            multiple = 'stack')
```

```python
# Box plots
fig = px.box(gapminder, x = 'lifeExp', color = 'continent')
fig
```
